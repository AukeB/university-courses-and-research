```python
# Imports
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# Path to your LOGS folder.
PATH = '/home/auke/Desktop/temp/Lab1/LOGS'
```

```python
# Function to make sure strings are ordered as 1,2,3,... and not 1,10,100,101,...
def natural_string_sort(arr, b1, b2):
    def check_if_int(var):
        try:
            int(var)
            return True
        except ValueError:
            return False
        
    arr_sort = []
    l = 0
    
    for i in range(len(arr)):
        for j in range(len(arr)):
            value = arr[j][b1:b2]
            int_bool = check_if_int(value)
            if int_bool:
                if int(arr[j][b1:b2]) == l+1:
                    arr_sort.append(arr[j])
            else:
                if int(arr[j][b1:b2-1]) == l+1:
                    arr_sort.append(arr[j])
            
        l += 1
        
    return arr_sort
```

```python
def create_h_file_list(path):
    # In here the filelist will be stored.
    files = []

    # Make an array that contains all the history files.
    for i in os.listdir(path):
        if os.path.isfile(os.path.join(path, i)) and ('1e_evolve_12ME_0.1' in i or '2c' in i): # Include all files in the folder that ...
            files.append(path + '/' + i) # ... contain the word history in it.

    files = sorted(files) # Sort the array.

    for i in range(len(files)):
        #print(files[i]) # <--- Toggle on to see if it works as intended.
        pass
    
    return files
```

```python
def create_p_file_list(path):
    # List of directories:
    dirs = []
    final_file_list = []
    
    for i in os.listdir(path):
        if os.path.isdir(path + '/' + i) and '12ME_0.1' in i:
            dirs.append(path + '/' + i)
                        
    for i in dirs:
        file_list = []
        files = os.listdir(i)
        for j in files:
            if 'data' in j:
                file_list.append(j)
                
        file_list = natural_string_sort(file_list, 7, -5)          
        final_file_list.append(i + '/' + file_list[-1]) # Only take the last profile file.
    
    return final_file_list
```

```python
def read_in_data_file(path):
    # First import the first part of the .data file, which is the global data.
    global_data = pd.read_csv(path, delim_whitespace=True, engine='python', nrows=3) # Read in first 3 rows.
    global_data.columns = global_data.iloc[0] # Use the second row as header.
    global_data = global_data.drop(global_data.index[0]) # Delete first row that only contains row numbers.

    # Secondly, import the time data, which is done less devious.
    time_data = pd.read_csv(path, delim_whitespace=True, engine='python', skiprows=5, header=0)
    
    return global_data, time_data
```

```python
# Create the filelist.
h_files = create_h_file_list(PATH)
p_files = create_p_file_list(PATH)

# Swap some elements to obtain the desired sequence.
h_files[1], h_files[2] = h_files[2], h_files[1]
p_files[0], p_files[2] = p_files[2], p_files[0]
p_files[1], p_files[2] = p_files[2], p_files[1]
```

## Global plot variables

```python
# Global figure variables
figure_size = (12,8)
fontsize_title = 20
fontsize_axis = 18

R_earth = 6.37e8 # Earth radius (cm)
R_sun = 6.9598e10 # Sun radius (cm)
R_jup = 6.9911e9 # Jupiter radius (cm)

labels = [
    r'Non-irradiated, $M = 12M_\oplus$, $f = 0.1$',
    r'$10 F_\oplus, M = 12M_\oplus$, $f = 0.1$',
    r'$100 F_\oplus, M = 12M_\oplus$, $f = 0.1$'
]

color_arr = ['black', 'chocolate', 'red']
linestyle_arr = ['-','-','-']
size_arr = [1,1,4]

adia = r'$\nabla_{ad}$'
rad = r'$\nabla_{rad}$'

sns.set_style('darkgrid', {'legend.frameon':True})
```

## Plot 1: The evolution of the radius of the 3 planets with time.

```python
plt.figure(figsize=(figure_size))

for i in range(len(h_files)):
    global_data, time_data = read_in_data_file(h_files[i])
    if i == 0:
        s = int(len(time_data['log_star_age'])/2) # Apparently the data is duplified.
        plt.plot(time_data['log_star_age'].iloc[0:s], time_data['radius'].iloc[0:s] * R_sun / R_earth, c=color_arr[i], linestyle=linestyle_arr[i], label=labels[i])
    else:
        plt.plot(time_data['log_star_age'], time_data['radius'] * R_sun / R_earth, c=color_arr[i], linestyle=linestyle_arr[i], label=labels[i])
    
legend = plt.legend(prop={'size': 15})
frame = legend.get_frame()
frame.set_color('lightgray')
frame.set_linewidth(10)
plt.title('The evolution of the radius of the 3 planets with time', fontsize = fontsize_title)
plt.xlabel(r'$\log$ Planet age [yr]', fontsize = fontsize_axis)
plt.ylabel('Radius [$R_\oplus$]', fontsize = fontsize_axis)
plt.savefig('fig5.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```

## Plot 2: The evolution of the luminosity of the 3 planets with time.

# Plot 3: The final mass (x-axis) vs. the final radius (y-axis) of all the 3 planets.

## Plot 4: Final radiative gradient and the adiabatic gradient  vs. the radius normalised to the total radius of the planet for the two extreme cases (the non irradiated one and the most irradiated one).

```python
plt.figure(figsize=(figure_size))

for i in range(len(p_files)):
    if i != 1:
        global_data, time_data = read_in_data_file(p_files[i])
        s = time_data['radius_cm'].iloc[0]

        plt.plot(time_data['radius_cm'] / s, time_data['grada'], 
                 c = color_arr[i], label = f'{labels[i]}, {adia}')
        plt.plot(time_data['radius_cm'] / s, time_data['gradr'], 
                 c = color_arr[i], linestyle='--', label = f'{labels[i]}, {rad}')
    
legend = plt.legend(loc = 'upper left')
frame = legend.get_frame()
frame.set_color('lightgray')
frame.set_linewidth(10)

plt.title('The final radiative and adiabatic gradient vs. the radius\nfor the 2 extreme cases', fontsize = fontsize_title)
plt.xlabel(r'Normalised planet radius', fontsize = fontsize_axis)
plt.ylabel(r'Gradients $\nabla$', fontsize = fontsize_axis)

plt.savefig('fig8.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```
