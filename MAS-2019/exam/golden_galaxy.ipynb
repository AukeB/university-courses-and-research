```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Global variables
phi = 1.686 
theta = np.linspace(0,2*np.pi,1000)
```

```python
# Computes coordinates from the angle.
def coordinates(theta,phi):
    # Computes radius from the angle.
    def radius_calc(theta,phi):
        return phi**theta
    radius = radius_calc(theta,phi)
    x1 = radius*np.cos(theta); y1 = radius*np.sin(theta)
    x2 = radius*np.cos(theta+np.pi); y2 = radius*np.sin(theta+np.pi)
    return x1,y1,x2,y2
```

```python
# Compute coordinates
x1,y1,x2,y2 = coordinates(theta,phi)
```

```python
# Plots just the analytical lines.
plt.rc('axes',labelsize=14)
plt.plot(x1,y1,c='b')
plt.plot(x2,y2,c='b')
plt.title('Framework')
plt.xlabel('x'); plt.ylabel('y')
plt.grid()
plt.show()
```

```python
# Generate stars next to lines.
x1_stars = np.random.normal(0,0.5,1000) + x1
y1_stars = np.random.normal(0,0.5,1000) + y1
x2_stars = np.random.normal(0,0.5,1000) + x2
y2_stars = np.random.normal(0,0.5,1000) + y2
```

```python
# Plot with stars.
plt.plot(x1,y1,c='k')
plt.plot(x2,y2,c='k')
plt.grid()
plt.scatter(x1_stars,y1_stars,s=0.5,c='b')
plt.scatter(x2_stars,y2_stars,s=0.5,c='b')
plt.xlabel('x');plt.ylabel('y')
plt.title('Stars added')
plt.show()
```
