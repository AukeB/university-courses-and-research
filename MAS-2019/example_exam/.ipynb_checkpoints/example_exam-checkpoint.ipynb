### 1 c) Define and explain the following quantities (2 CP)
1. Posterior (0.5 CP)
2. Likelihood (0.5 CP)
3. Prior (0.5 CP)
5. Burnin (0.5 CP)

**Answer**:
- **Posterior**: The posterior probability distribution is the conditional probability that is assigned after the relevant evidence is taken into account, contrary to the prior probability distribution. It describes the human attempt of inferring the parameter values $\theta$ from observed data $\vec{x}$.
- **Likelihood**: The Likelihood function is the probability function of observed data expressed as function of observed data. It describes the order in which the noise process occurs in nature.
- **Prior**: The prior (prior probability distribution) is the probability distribution that would express one's beliefs about this quantity before some evidence is taken into account. To relate the posterior and the likelihood, *priors* are required. They describe an a-priori  probability of data $\vec{x}$ or parameter values $\theta$.
- **Burnin**: Burn-in is a part of the Markov Chain Monte Carlo (MCMC), which allows you to find a starting point. It describes the practice of throwing away some interations at the beginning of an MCMC run. The part you throw away is called the burn-in period.

### 1 e) Normalization (1 CP)
if $P(x)$ is said to be normalized, with respect to which variable is it then normalized? Which equation describes 'normalization'? Finally, if $P(x)$ is normalized, is $P(\theta|x)$ normalized?

If $\mathcal{P}(x)$ is normalized, it is normalized w.r.t. the variable $x$.

The equation that describes the normalization is $ \frac{1}{A} \int \mathcal{P}(x) dx = 1 $, with $A$ as normalization constant.

If $\mathcal{P}(x)$ is normalized, it does not mean that $\mathcal{P}(\theta|x)$ is also normalized, because it is the probability function of another variable, $\theta$, and the relation between $x$ and $\theta$ is not specified.

1. Read in the chain (0 CP)

```python
# Initialisations.
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read in the data file with np.loadtxt.
markov_chain = np.loadtxt('MarkovChain.txt',dtype=float,comments='#',delimiter=None,skiprows=0,usecols=None)
print(f'Shape of data: {markov_chain.shape}') # Print shape of the data.
```

```python
# Same for the second column ...
plt.plot(number_array,markov_chain[:,1])
plt.title('Column 1: Radius')
plt.xlabel('Iterations'); plt.ylabel('radius $r$ [kpc]')
plt.grid()
plt.show()
```

```python
# ... and third column.
plt.plot(number_array,markov_chain[:,2])
plt.title('Column 2: $\log$ likelihood')
plt.xlabel('Iterations'); plt.ylabel('$\log \mathcal{L}$')
plt.grid()
plt.show()
```

```python
# Plotting the mass versus the radius, with the $\log$ likelihood as color map.
plt.plot(markov_chain[:,0],markov_chain[:,1],lw=0.5,c='k')
plt.title('Mass vs. radius')
plt.xlabel('Mass $M$ $[M_{Sun}]$'); plt.ylabel('radius $r$ [kpc]')
plt.scatter(markov_chain[:,0],markov_chain[:,1],c=np.log10(markov_chain[:,2]))
cbar = plt.colorbar()
cbar.set_label('$\log_{10}(\log \mathcal{L})$')
plt.grid()
plt.show()
```

### 3 b) Sampling (2 CP)

- Generate 5000 random samples from a uniform distribution with limits [0,1] (0.5 CP)

```python
random_samples = np.random.uniform(0,1,5000)
```

```python
plt.rc('axes',labelsize=14)
plt.hist(random_samples,bins=10,rwidth=0.9,color='gray')
plt.xlabel('$x$');plt.ylabel('counts')
plt.xlim(-1,1)
plt.title('5000 samples from a uniform distribution')
plt.show()
plt.close()
```

- For the remainder of this exercise, do not use a random number generator anymore. Instead work with the samples you already have.
    - Transform your samples to now sample from a Uniform distribution with limits [-1,+1] (0.5 CP)
    - Use your samples to generate samples from a triangular distribution (1 CP)

```python
new_samples = []
for i in range(len(random_samples)):
    dif = 1-random_samples[i]
    new_samples.append(1-2*dif)
```

```python
plt.hist(new_samples,bins=10,rwidth=0.9,color='gray')
plt.xlabel('$x$');plt.ylabel('counts')
plt.title('5000 samples from [-1,1] distribution')
plt.show()
plt.close()
```

```python
triangular = []
for i in range(len(new_samples)):
    for j in range(len(new_samples)):
        triangular.append(new_samples[j]+new_samples[i])
np.random.shuffle(triangular)
```

```python
plt.hist(triangular[::5000],bins=11,rwidth=0.9,color='gray',density=True)
plt.xlabel('$x$');plt.ylabel('counts')
plt.title('5000 samples from a triangular distribution')
plt.show()
plt.close()
```
